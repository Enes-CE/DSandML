```python
# independent(x) and dependent(y) features (X genelde byk, y genelde kk harfle gsterilir)
X =df[["Study Hours", "Sleep Hours", "Attendance Rate", "Social Media Hours"]]
y = df["Exam Score"]
```

```python
X.head()
```

```python
y.head()
```

```python
# Train - Test Split (Train eitimde kullanmak ve modeli oluturmak iin, Test modeli test etmek iin)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 15)
```

```python
# standardize the data set
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
X_test[0]
```

```python
df.iloc[0]
```

```python
new_student = [[5, 7, 90, 2]]
```

```python
new_student_scaled = scaler.transform(new_student)
```

```python
reg.predict(new_student_scaled)
```

```python
# prediction
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("mse: ", mse)
print("mae: ", mae)
print("rmse: ", rmse)
```

```python
r2 = r2_score(y_test, y_pred)
print("r2 score: ", r2)
```

```python
#adjusted R-squared
print(1 - (1-r2)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1))
```
